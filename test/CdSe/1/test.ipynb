```python
from pyscf import gto, scf, dft
from pyscf.geomopt.berny_solver import optimize

mol = gto.M(atom='''
 Se   -2.0118679    0.0000000   -0.0000000
 Cd   -0.0000000   -1.6205108    0.0000000
 Se    2.0118679   -0.0000000    0.0000000
 Cd    0.0000000    1.6205108   -0.0000000
''', basis='def2-tzvp', ecp='lanl2dz')
mf = dft.RKS(mol)
mf.xc = 'pbe'
mf.max_cycle = 500
mf.verbose = 4

# #
# # geometry optimization for HF.  There are two entries to invoke the berny
# # geometry optimization.
# #
# # method 1: import the optimize function from pyscf.geomopt.berny_solver
mol_eq = optimize(mf, maxsteps=200)
print(mol_eq.atom_coords())

# method 2: create the optimizer from Gradients class
# mol_eq = mf.Gradients().optimizer(solver='geometric', maxsteps=200).kernel()
```
